```python
import time

def translate(string):
    try:
        time.sleep(0.2)
        translated = translator.detect(string).lang
        return translated
    except:
        pass
        
```

```python
spam400 = spam400['message'].to_list()
```

```python
translated = []
for element in spam400:
    translated.append(translate(element))
    
```
